```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
file_path_1 = 'D:\FREELANCE_PROJECTS\product-recommendation\data\All Electronics.csv'
file_path_2 = 'D:\FREELANCE_PROJECTS\product-recommendation\data\Air Conditioners (1).csv'

all_electronics_df = pd.read_csv(file_path_1)
air_conditioners_df = pd.read_csv(file_path_2)
```

```python
combined_df = pd.concat([all_electronics_df, air_conditioners_df], ignore_index=True)
```

```python
combined_df.tail()
```

```python
combined_df.fillna("", inplace=True)
```

```python
combined_df['text_features'] = combined_df['name'] + " " + combined_df['sub_category']
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
```

```python
tfidf_matrix = tfidf_vectorizer.fit_transform(combined_df['text_features'])
```

```python
cosine_sim_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def recommend_products(product_name, top_n=5):
    idx = combined_df[combined_df['name'].str.contains(product_name, case=False, na=False)].index
    if idx.empty:
        return "Product not found."
    
    sim_scores = list(enumerate(cosine_sim_matrix[idx[0]]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    recommended_indices = [i[0] for i in sim_scores[1:top_n + 1]]
    
    return combined_df.iloc[recommended_indices][['name', 'sub_category', 'discount_price']]

sample_recommendations = recommend_products("Redmi 10")
sample_recommendations
```

```python
def precision_at_k(recommended_items, relevant_items, k):
    recommended_top_k = recommended_items[:k]
    relevant_set = set(relevant_items)
    recommended_set = set(recommended_top_k)
    return len(recommended_set & relevant_set) / k


def recall_at_k(recommended_items, relevant_items, k):
    recommended_top_k = recommended_items[:k]
    relevant_set = set(relevant_items)
    recommended_set = set(recommended_top_k)
    return len(recommended_set & relevant_set) / len(relevant_set)

def mean_reciprocal_rank(recommended_items, relevant_items):
    for idx, item in enumerate(recommended_items):
        if item in relevant_items:
            return 1 / (idx + 1)
    return 0

def ndcg(recommended_items, relevant_items, k):
    dcg = sum([1 / np.log2(idx + 2) if recommended_items[idx] in relevant_items else 0 for idx in range(k)])
    idcg = sum([1 / np.log2(idx + 2) for idx in range(min(len(relevant_items), k))])
    return dcg / idcg if idcg > 0 else 0

def coverage(recommended_items, catalog):
    return len(set(recommended_items)) / len(catalog)

def diversity(recommended_items, product_metadata):
    sub_categories = [product_metadata[item]['sub_category'] for item in recommended_items]
    return len(set(sub_categories)) / len(sub_categories)
```

```python
import numpy as np
recommended_items = ["Redmi 10", "Samsung Galaxy", "iPhone 12", "OnePlus Nord", "Moto G"]
relevant_items = ["iPhone 12", "Samsung Galaxy", "Moto G"]
catalog = ["Redmi 10", "Samsung Galaxy", "iPhone 12", "OnePlus Nord", "Moto G", "Realme C55", "Vivo V27"]

# Metrics
p_at_k = precision_at_k(recommended_items, relevant_items, 3)
r_at_k = recall_at_k(recommended_items, relevant_items, 3)
mrr = mean_reciprocal_rank(recommended_items, relevant_items)
ndcg_score = ndcg(recommended_items, relevant_items, 3)
coverage_score = coverage(recommended_items, catalog)
```
